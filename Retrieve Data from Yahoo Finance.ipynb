```python
import pandas as pd
import numpy as np
from pandas_datareader import data
import datetime
import zipfile
import os
```

```python
sp500_list=pd.read_csv("SP-500-Index-Constitents-List-Feb-2018.csv",index_col="S.No.")
```

```python
start=datetime.datetime(1990,1,1)
end=datetime.datetime(2018,2,28)
save_directory="./historical_data/"
error_list=[]
for ticker in sp500_list["Ticker"].values:
    print("Retrieving {}'s historical data...".format(ticker))
    try:
        hist_data=data.get_data_yahoo(ticker,start,end)
        start_date=str(hist_data.index[0])[:10]
        end_date=str(hist_data.index[-1])[:10]
        hist_data.to_csv(save_directory+ticker+"+"+start_date+"+"+end_date+".csv")
        print("Successfully retrieve {}'s historical data!".format(ticker))
    except:
        print("Something wrong happens...")
        error_list.append(ticker)  
```

```python
error_list_2=[]
for ticker in error_list:
    print("Retrieving {}'s historical data...".format(ticker))
    try:
        hist_data=data.get_data_yahoo(ticker,start,end)
        start_date=str(hist_data.index[0])[:10]
        end_date=str(hist_data.index[-1])[:10]
        hist_data.to_csv(save_directory+ticker+"+"+start_date+"+"+end_date+".csv")
        print("Successfully retrieve {}'s historical data!".format(ticker))
    except:
        print("Something wrong happens...")
        error_list_2.append(ticker)   
```

```python
error_list_3=[]
for ticker in error_list_2:
    print("Retrieving {}'s historical data...".format(ticker))
    try:
        hist_data=data.get_data_yahoo(ticker,start,end)
        start_date=str(hist_data.index[0])[:10]
        end_date=str(hist_data.index[-1])[:10]
        hist_data.to_csv(save_directory+ticker+"+"+start_date+"+"+end_date+".csv")
        print("Successfully retrieve {}'s historical data!".format(ticker))
    except:
        print("Something wrong happens...")
        error_list_3.append(ticker)   
```

```python
save_directory="./historical_data/"
zip_obj=zipfile.ZipFile(save_directory+"stock_price.zip","w",zipfile.ZIP_DEFLATED)

for filename in os.listdir(save_directory):
    if filename in [".DS_Store","stock_price.zip"]:
        continue
    zip_obj.write(save_directory+filename)

zip_obj.close()
```
